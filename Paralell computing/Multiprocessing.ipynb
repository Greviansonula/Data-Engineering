```python
import time
from multiprocessing import Pool

# Define a function to be executed serially or in parallel
def my_function(x):
    time.sleep(1)
    return x**2

# Create a list of input data
input_data = [1, 2, 3, 4, 5]

# Perform serial processing
start_time = time.time()
result = [my_function(x) for x in input_data]
print("Serial processing time:", time.time() - start_time)

# Perform parallel processing with 4 worker processes
with Pool(4) as p:
    start_time = time.time()
    result = p.map(my_function, input_data)
    print("Parallel processing time:", time.time() - start_time)
```
